```python
# -*- coding: utf-8 -*-
"""

"""


import os
import cv2
import random
import numpy as np

def Crop(im, Rangox,RangoY):
    normal = im[RangoY[0]:RangoY[1], Rangox[0]:Rangox[1]]
    #print(Rangox," ",RangoY)
    #Los modelos de AdapNet++ estan entrenados para recibir imágenes de tamaño 768x384 como entrada.
    normal=cv2.resize(normal,(768,384))
    return normal

def Scale(im,scale):
    h_new=im.shape[0]*scale
    w_new=im.shape[1]*scale
    
    im_scale=cv2.resize(im,(w_new,h_new))
    Rangoy=np.uint16([(h_new/2)-192,(h_new/2)+192])
    Rangox=np.uint16([(w_new/2)-384,(w_new/2)+384])
   
    im_scale=im_scale[Rangoy[0]:Rangoy[1],Rangox[0]:Rangox[1]]
    
    return im_scale

def Flip(im):
    #Flip horizontal-->Eje 1
    flippedimage= cv2.flip(im, 1)
    flippedimage=cv2.resize(flippedimage,(768,384))
    return flippedimage

def Resize(im):
    normal=cv2.resize(im,(768,384))
    return normal

def main():
    Dir_RGB='Introducir dirección origen imágenes RGB'
    Dir_ther='Introducir dirección origen imágenes térmicas'
    Dir_labels='Introducir dirección origen etiquetas'
    
    Dest_rgb='Introducir dirección destino imágenes RGB'
    Dest_ther='Introducir dirección destino imágenes térmicas'
    Dest_lab='Introducir dirección destino etiquetas'
    #Dest_resize_rgb='C:\\Users\\Usuario\\Desktop\\Images-rescate\\Etiquetas\\RGB\\Rescate4\\RGB_RESIZE'
    #Dest_resize_label='C:\\Users\\Usuario\\Desktop\\Images-rescate\\Etiquetas\\RGB\\Rescate4\\LABELS_RESIZE'
    #Parte final del nombre correspondiente a cada transformación
    ext_crop='_crop.png'
    ext_flip='_flip.png'
    ext_sc='_scale.png'
    ext_res='_resize.png'
    #Listas con los nombres de las imágenes de cada tipo
    Lista_RGB=os.listdir(Dir_RGB)
    Lista_labels=os.listdir(Dir_labels)
    Lista_Ther=os.listdir(Dir_ther)
    for i in range(len(Lista_RGB)):
        x1_rand=random.randint(0,500)
        x2_rand=random.randint(x1_rand+100,704)
        
        y1_rand=random.randint(0,400)
        y2_rand=random.randint(y1_rand+100,576)
        
        Rangox=[x1_rand,x2_rand]
        RangoY=[y1_rand,y2_rand]
        
        scale=random.randint(2,4)
        #Imágen térmica, RGB y etiqueta deben coincidir. Reciben exactamente la misma transformación.
        im_rgb=cv2.imread(os.path.join(Dir_RGB,Lista_RGB[i]))
        im_ther=cv2.imread(os.path.join(Dir_ther,Lista_Ther[i]))
        im_lab=cv2.imread(os.path.join(Dir_labels,Lista_labels[i]),cv2.IMREAD_UNCHANGED)
        
        im_rgb_crop=Crop(im_rgb,Rangox,RangoY)
        im_lab_crop=Crop(im_lab,Rangox,RangoY)
        im_ther_crop=Crop(im_ther,Rangox,RangoY)
        
        im_rgb_sc=Scale(im_rgb,scale)
        im_lab_sc=Scale(im_lab,scale)
        im_ther_sc=Scale(im_ther,scale)
        
        im_rgb_flip=Flip(im_rgb)
        im_lab_flip=Flip(im_lab)
        im_ther_flip=Flip(im_ther)
        
        im_rgb_res=Resize(im_rgb)
        im_lab_res=Resize(im_lab)
        im_ther_res=Resize(im_ther)
        #Se crean los nombres para las nuevas imágenes
        name1_rgb=Lista_RGB[i].split('.jpg')
        #print(name1_rgb)
        name1_rgb=name1_rgb[0]
        name1_lab=Lista_labels[i].split('.png')
        name1_lab=name1_lab[0]
        name1_ther=Lista_Ther[i].split('.jpg')
        name1_ther=name1_ther[0]
        
        name1_rgb=os.path.join(Dest_rgb,name1_rgb)
        name_crop_rgb="".join((name1_rgb,ext_crop))
        name_flip_rgb="".join((name1_rgb,ext_flip))
        name_sc_rgb="".join((name1_rgb,ext_sc))
        name_res_rgb="".join((name1_rgb,ext_res))
        
        #print(name_crop_rgb)
        name1_lab=os.path.join(Dest_lab,name1_lab)
        name_crop_lab="".join((name1_lab,ext_crop))
        name_flip_lab="".join((name1_lab,ext_flip))
        name_sc_lab="".join((name1_lab,ext_sc))
        name_res_lab="".join((name1_lab,ext_res))
        
        name1_ther=os.path.join(Dest_ther,name1_ther)
        name_crop_ther="".join((name1_ther,ext_crop))
        name_flip_ther="".join((name1_ther,ext_flip))
        name_sc_ther="".join((name1_ther,ext_sc))
        name_res_ther="".join((name1_ther,ext_res))
        #Se guardan las imágenes
        cv2.imwrite(name_crop_rgb,im_rgb_crop)
        cv2.imwrite(name_flip_rgb,im_rgb_flip)
        cv2.imwrite(name_sc_rgb,im_rgb_sc)
        cv2.imwrite(name_res_rgb,im_rgb_res)
        
        cv2.imwrite(name_crop_lab,im_lab_crop)
        cv2.imwrite(name_flip_lab,im_lab_flip)
        cv2.imwrite(name_sc_lab,im_lab_sc)
        cv2.imwrite(name_res_lab,im_lab_res)
        
        cv2.imwrite(name_crop_ther,im_ther_crop)
        cv2.imwrite(name_flip_ther,im_ther_flip)
        cv2.imwrite(name_sc_ther,im_ther_sc)
        cv2.imwrite(name_res_ther,im_ther_res)
        
if __name__ == '__main__':
    main()
```
