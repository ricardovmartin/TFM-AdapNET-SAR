```python
"""
El parámetro FOV de las cámaras térmica y RGB utilizadas para tomar las fotos del dataset UMA-SAR no es el mismo.
Hay que manipular las imágenes RGB y que coincidan con las térmicas. 
"""

import os 
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

sc_w=math.tan(math.radians(57.8))/math.tan(math.radians(44))
sc_h=math.tan(math.radians(47.3))/math.tan(math.radians(36))

Dir='Dirección imágenes RGB'
Dir2='Dirección imágenes térmicas'
Lista1=os.listdir(Dir)
Lista2=os.listdir(Dir2)
Dest1='Dirección en la que se guardarán las imágenes RGB transformadas'
Dest2='Dirección en la que se guardarán las imágenes térmicas transformadas'


def rotate(image,angle):
    image_center = tuple(np.array(image.shape[1::-1]) / 2)
    rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
    result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
    return result


def scale(im):
    h_new = np.uint16(576*sc_h)
    w_new = np.uint16(704*sc_w)
    new_shape = (w_new,h_new)
    im_re_scale=cv2.resize(im,new_shape,interpolation = cv2.INTER_AREA)
    Rangox=np.uint16([(w_new/2)-365,(w_new/2)+380])
    Rangoy=np.uint16([(h_new/2)-305,(h_new/2)+350])
    print(Rangox,' ',Rangoy)
    im_rs_def=im_re_scale[Rangoy[0]:Rangoy[1],Rangox[0]:Rangox[1]]
    im_rs_def = cv2.resize(im_rs_def, (704, 576),interpolation = cv2.INTER_AREA)
    
    return im_rs_def

def scale_the(im):
    rangox_the=np.uint16([(im.shape[1]/2)-295,(im.shape[1]/2)+290])
    im=im[0:575,rangox_the[0]:rangox_the[1]]
    im1_the = cv2.resize(im, (704, 576),interpolation = cv2.INTER_AREA)
    return im1_the
    
    

for i in range(len(Lista1)):
    path1=os.path.join(Dir,Lista1[i])
    path2=os.path.join(Dir2,Lista2[i])
    im_rgb=cv2.imread(path1)
    im_the=cv2.imread(path2)
    
    im_rot=rotate(im_rgb,-3)
    im_rgb_sc=scale(im_rot)

    im_the_sc=scale_the(im_the)
    
    cv2.imwrite(os.path.join(Dest1,Lista1[i]),im_rgb_sc)
    cv2.imwrite(os.path.join(Dest2,Lista1[i]),im_the_sc)
```
